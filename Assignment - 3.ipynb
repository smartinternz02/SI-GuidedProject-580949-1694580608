```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Datasets/car_crashes.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='speeding', y='alcohol', data=df)
```

```python
sns.scatterplot(x='speeding', y='not_distracted', data=df)
```

```python
sns.lineplot(x='speeding', y='alcohol', data=df)
```

```python
sns.lineplot(x='speeding', y='not_distracted', data=df)
```

```python
sns.distplot(df['speeding'])
```

```python
sns.relplot(x='speeding', y='alcohol', data=df, hue='not_distracted')
```

```python
plt.subplots(figsize=(20,15))
sns.barplot(y='abbrev', x='alcohol', data=df)
```

```python
plt.subplots(figsize=(20,15))
sns.barplot(x='total', y='speeding', data=df)
```

```python
sns.jointplot(x='total', y='speeding', data=df)
```

```python
corr = df.corr()
```

```python
sns.heatmap(corr, annot=True, cmap="YlGnBu")
```
